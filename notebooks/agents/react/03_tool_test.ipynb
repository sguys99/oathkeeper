#### 1-1. calculate_roi вҖ” мҲҳмқөлҘ /л§Ҳм§„ кі„мӮ°

```python
from backend.app.agent.tools import calculate_roi

# Case 1: м •мғҒ мҲҳмқө (л§Ҳм§„ 30%+)
result = await calculate_roi.ainvoke({
    "contract_amount": 20000,  # 2м–өмӣҗ
    "estimated_cost": 13000,   # 1.3м–өмӣҗ
    "duration_months": 5.0,
})
roi = json.loads(result)
print("=== Case 1: кі л§Ҳм§„ н”„лЎңм қнҠё ===")
print(json.dumps(roi, ensure_ascii=False, indent=2))
print(f"л§Ҳм§„мңЁ: {roi['margin_pct']}%, л“ұкёү: {roi['profitability_rating']}")
```

```python
# Case 2: мҶҗмӢӨ н”„лЎңм қнҠё (л§Ҳм§„ мқҢмҲҳ)
result = await calculate_roi.ainvoke({
    "contract_amount": 8000,
    "estimated_cost": 10000,
    "duration_months": 6.0,
})
roi_loss = json.loads(result)
print(f"л§Ҳм§„мңЁ: {roi_loss['margin_pct']}%, л“ұкёү: {roi_loss['profitability_rating']}")
assert roi_loss['profitability_rating'] == 'loss', "мқҢмҲҳ л§Ҳм§„ вҶ’ loss л“ұкёүмқҙм–ҙм•ј н•Ё"
```

#### 1-2. calculate_weighted_score вҖ” к°ҖмӨ‘ м җмҲҳ мһ¬кі„мӮ°

```python
from backend.app.agent.tools import calculate_weighted_score

# 7к°ң кё°мӨҖ м җмҲҳ мһ…л Ҙ
sample_scores = [
    {"criterion": "кё°мҲ  м Ғн•©м„ұ", "score": 80, "weight": 0.20},
    {"criterion": "мҲҳмқөм„ұ", "score": 65, "weight": 0.20},
    {"criterion": "лҰ¬мҶҢмҠӨ к°Җмҡ©м„ұ", "score": 70, "weight": 0.15},
    {"criterion": "лӮ©кё° лҰ¬мҠӨнҒ¬", "score": 60, "weight": 0.10},
    {"criterion": "мҡ”кө¬мӮ¬н•ӯ лӘ…нҷ•м„ұ", "score": 75, "weight": 0.10},
    {"criterion": "м „лһөм Ғ к°Җм№ҳ", "score": 85, "weight": 0.15},
    {"criterion": "кі к°қ лҰ¬мҠӨнҒ¬", "score": 70, "weight": 0.10},
]

result = await calculate_weighted_score.ainvoke({"scores": sample_scores})
ws = json.loads(result)
print(json.dumps(ws, ensure_ascii=False, indent=2))
print(f"\nмҙқм җ: {ws['total_score']}, нҢҗм •: {ws['verdict']}, к°ҖмӨ‘м№ҳ н•©: {ws['weight_sum']}")
```

```python
# Verdict мһ„кі„к°’ кІҖмҰқ
for score_val, expected_verdict in [(90, "go"), (70, "go"), (55, "conditional_go"), (40, "conditional_go"), (30, "no_go")]:
    test_scores = [{"criterion": "test", "score": score_val, "weight": 1.0}]
    result = await calculate_weighted_score.ainvoke({"scores": test_scores})
    v = json.loads(result)["verdict"]
    status = "вң“" if v == expected_verdict else "вң—"
    print(f"  {status} score={score_val} вҶ’ {v} (expected: {expected_verdict})")
```

#### 1-3. estimate_timeline вҖ” PERT кё°л°ҳ мқјм • м¶”м •

```python
from backend.app.agent.tools import estimate_timeline

# Case 1: ліөмһЎлҸ„ лҶ’мқҢ, лҢҖк·ңлӘЁ нҢҖ, мҷёл¶Җ мқҳмЎҙм„ұ мһҲмқҢ
result = await estimate_timeline.ainvoke({
    "base_duration_months": 5.0,
    "team_size": 7,
    "complexity": "high",
    "has_external_dependencies": True,
})
tl = json.loads(result)
print("=== High complexity, large team, external deps ===")
print(json.dumps(tl, ensure_ascii=False, indent=2))
print(f"\nPERT м¶”м •: {tl['pert_estimate_months']}к°ңмӣ” + лІ„нҚј {tl['buffer_months']}к°ңмӣ” = к¶ҢмһҘ {tl['recommended_months']}к°ңмӣ”")
```

```python
# Case 2: лӢЁмҲң н”„лЎңм қнҠё
result = await estimate_timeline.ainvoke({
    "base_duration_months": 3.0,
    "team_size": 3,
    "complexity": "low",
    "has_external_dependencies": False,
})
tl_simple = json.loads(result)
print(f"лӢЁмҲң н”„лЎңм қнҠё: base=3к°ңмӣ” вҶ’ к¶ҢмһҘ {tl_simple['recommended_months']}к°ңмӣ”")
assert tl_simple['recommended_months'] < tl['recommended_months'], "лӢЁмҲң н”„лЎңм қнҠёк°Җ лҚ” м§§м•„м•ј н•Ё"
```

#### 1-4. assess_risk_matrix вҖ” нҷ•лҘ  Г— мҳҒн–ҘлҸ„ л§ӨнҠёлҰӯмҠӨ

```python
from backend.app.agent.tools import assess_risk_matrix

sample_risks = [
    {"category": "кё°мҲ ", "item": "мӢ к·ң кё°мҲ  лҸ„мһ… лҰ¬мҠӨнҒ¬", "probability": "HIGH", "impact": "HIGH"},
    {"category": "мқјм •", "item": "мқјм • мҙҲкіј лҰ¬мҠӨнҒ¬", "probability": "MEDIUM", "impact": "HIGH"},
    {"category": "мһ¬л¬ҙ", "item": "л№„мҡ© мҙҲкіј лҰ¬мҠӨнҒ¬", "probability": "LOW", "impact": "MEDIUM"},
    {"category": "кі к°қ", "item": "мҡ”кө¬мӮ¬н•ӯ ліҖкІҪ", "probability": "MEDIUM", "impact": "MEDIUM"},
    {"category": "лІ”мң„", "item": "лІ”мң„ нҷ•мһҘ", "probability": "HIGH", "impact": "LOW"},
]

result = await assess_risk_matrix.ainvoke({"risks": sample_risks})
rm = json.loads(result)
print(json.dumps(rm, ensure_ascii=False, indent=2))

# level кІҖмҰқ: HIGHГ—HIGH вҶ’ CRITICAL
first_risk = rm['risks'][0]
assert first_risk['level'] == 'CRITICAL', f"HIGHГ—HIGHлҠ” CRITICALмқҙм–ҙм•ј н•Ё, got: {first_risk['level']}"
print(f"\nвң“ HIGHГ—HIGH вҶ’ {first_risk['level']}")
print(f"м „мІҙ лҰ¬мҠӨнҒ¬ н”„лЎңнҢҢмқј: {rm['summary']['overall_risk_profile']}")
```
